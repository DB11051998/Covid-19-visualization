```python
## as there was no coordinate in the scraped dataset
## so i had to scrape the data from the below taken address
## while scraping i did'nt get the all states coordinates
info = pd.read_html("http://www.quickgs.com/latitudinal-and-longitudinal-extents-of-india-indian-states-and-cities/") 

#convering the table data into DataFrame
coordinates = pd.DataFrame(info[0])

coordinates['state']=coordinates['State']
```

```python
## function to remove the N and E 
def data_pre(cord):
    return cord[0:5]
```

```python
x=data.merge(coordinates,on='state',how='inner')
```

```python
x['Latitude']=x['Latitude'].apply(data_pre)## used pandas apply fuction to parse it
x['Longitude']=x['Longitude'].apply(data_pre)## used pandas apply fuction to parse it
```

```python
## converted to float type 
## as folium accepts float type latitude and longitude while marking on the map
x['Latitude']=x['Latitude'].astype('float')
x['Longitude']=x['Longitude'].astype('float')
```

```python
x.drop(columns=['State'],inplace=True)
```

```python
x.head()
```

```python
statesnotpresent = list(set(data['state'][:-1].values)-set(coordinates['state'].values))
```

```python
## these are the states were n't present while
statesnotpresent
#{'chandigarh':[30.7333,76.7794],'Jammu and Kashmir':[33.7782,76.5762],'Telengana':[18.1124,79.0193],'Delhi':[28.7041,77.1025],'Puducheery':[11.9416,79.8083],'Andaman and Nicobar Islands':[11.7401,92.6586],'Dadra and Nagar Haveli and Daman and Diu':[20.4283,72.8397],'Ladakh':[34.152588,77.577049.]}
```

```python

states_not_data1=pd.DataFrame({'state':statesnotpresent,'Latitude':[30.7333,33.7782,18.1124,28.7041,11.9416,11.7401,20.4283,34.152588],'Longitude':[76.7794,76.5762,79.0193,77.1025,79.8083,92.6586,72.8397,77.577049]})
```

```python
X=data.merge(states_not_data1,on='state',how='inner').append(x, ignore_index=True)
```

```python
X['Deaths_cumulative_colors']=pd.cut(X['Deaths_cumulative'], bins=4,labels=['yellow', 'green', 'blue', 'red'])
```

```python
X['active_total_colors']=pd.cut(X['active_total'], bins=4,labels=['yellow', 'green', 'blue', 'red'])
```

## map with the bubble marking, which could be interpreted as the hotspot

Larger the radius of the bubble/circle more is the affeceted the place is

```python

# retreiving the data from final table and plotting it on the INDIA map
# creating the map object zooming on INDIA, location here shows the lat and long of INDIA
india = folium.Map(location = [20.5937,78.9629],zoom_start=4.5)
#adding to map
for state,lat,longi,total_cases,Death,Recov,color in zip(list(X['state']),list(X['Latitude']),list(X['Longitude']),list(X['active_total']),list(X['Cured/Discharged/Migrated_cumulative']),list(X['Deaths_cumulative']),list(X['active_total_colors'])):
    #for creating circle marker
    folium.CircleMarker(location = [lat,longi],radius = 5,color=color,fill = True,fill_color="color").add_to(india)
    folium.Circle(
      location=[lat,longi],
      radius=total_cases*2.5,
      color='red',
      fill=True,
      fill_color='red'
   ).add_to(india)

    #for creating marker
    folium.Marker(location = [lat,longi],
                popup='<strong><b>State  : '+state+'</strong> <br>' +
                      '<strong><b>Total Cases :'+str(total_cases)+'</striong><br>' +
                      '<strong><b>Deaths :'+str(Death)+'</striong><br>' +
                      '<strong><b>Recoveries :'+str(Recov)+'</striong><br>').add_to(india)
    #to show the map
india
## on pressing on the marker you will get all the stats
```

Stacked bar plot showing deaths and cured cases

```python

#Read in data & create total column
#stacked_bar_data = pd.read_csv("C:\stacked_bar.csv")
X["total"] = X['Cured/Discharged/Migrated_cumulative'] + X.Deaths_cumulative

#Set general plot properties
sns.set_style("white")
sns.set_context({"figure.figsize": (24, 10)})

#Plot 1 - background - "total" (top) series
sns.barplot(x = X.state, y = X.total, color = "green")

#Plot 2 - overlay - "bottom" series
bottom_plot = sns.barplot(x = X.state, y = X.Deaths_cumulative, color = 'red')


topbar = plt.Rectangle((0,0),1,1,fc="green", edgecolor = 'none')
bottombar = plt.Rectangle((0,0),1,1,fc='red',  edgecolor = 'none')
l = plt.legend([bottombar, topbar], ['Deaths', 'cured'], loc=1, ncol = 2, prop={'size':16})
l.draw_frame(False)

#Optional code - Make plot look nicer
sns.despine(left=True)
bottom_plot.set_ylabel("Y-axis label")
bottom_plot.set_xlabel("X-axis label")

#Set fonts to consistent 16pt size
for item in ([bottom_plot.xaxis.label, bottom_plot.yaxis.label] +
             bottom_plot.get_xticklabels() + bottom_plot.get_yticklabels()):
    item.set_fontsize(15)
plt.xticks(rotation=45)
```
